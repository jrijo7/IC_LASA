```python
import pandas as pd
import glob
import os
```

```python
# Definido o caminho dos arquivos CSV

caminho = "C:/Users/joaov/Documents/dados_saude/"
if os.path.exists(caminho):
  arquivos = os.listdir(caminho)
else:
  print(f'Erro: O diretório {caminho} não existe.')
```

```python
# Listar os arquivos CSV
arquivos_csv = glob.glob(os.path.join(caminho, "*.csv"))

# Carregar os arquivos no Dataframe

dfs = []

for arquivo in arquivos_csv:
  df = pd.read_csv(arquivo, low_memory=False)[['ocor_MUNNOME','TIPOBITO','data_obito','idade_obito_calculado','SEXO','RACACOR','ESC','OCUP','CIRCOBITO','def_circ_obito','ACIDTRAB', 'causabas_capitulo', 'causabas_grupo', 'causabas_categoria', 'causabas_subcategoria']]
  dfs.append(df)
  print(f'Lido {arquivo}')

df_final = pd.concat(dfs, ignore_index=True)
```

Opção 2 feita com: .size()

```python
df_agrupado_2 = df_final.groupby(by=['ocor_MUNNOME', 'data_obito']).size().reset_index(name='obitos_totais')
```

```python
df_agrupado_2.head()
```

```python
# Vamos ordenar os dados por município e data de óbito
df_agrupado_2 = df_agrupado_2.sort_values(by=['ocor_MUNNOME', 'data_obito'])

# Em seguida iremos calcular os óbitos esperados
df_agrupado_2['obitos_esperados'] = df_agrupado_2.groupby('ocor_MUNNOME')['obitos_totais'].transform(lambda x: x.rolling(30, min_periods=1).mean())
```
