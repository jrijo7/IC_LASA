```python
import pandas as pd
import glob
import os
```

```python
# Definido o caminho dos arquivos CSV

caminho = "C:/Users/joaov/Documents/dados_saude/"
if os.path.exists(caminho):
  arquivos = os.listdir(caminho)
else:
  print(f'Erro: O diretório {caminho} não existe.')
```

```python
# Listar os arquivos CSV
arquivos_csv = glob.glob(os.path.join(caminho, "*.csv"))

# Carregar os arquivos no Dataframe

dfs = []

for arquivo in arquivos_csv:
  df = pd.read_csv(arquivo, low_memory=False)[['ocor_MUNNOME','TIPOBITO','data_obito','idade_obito_calculado','SEXO','RACACOR','ESC','OCUP','CIRCOBITO','def_circ_obito','ACIDTRAB', 'causabas_capitulo', 'causabas_grupo', 'causabas_categoria', 'causabas_subcategoria']]
  dfs.append(df)
  print(f'Lido {arquivo}')

df_final = pd.concat(dfs, ignore_index=True)
```

Opção 2 feita com: .size()

```python
df_agrupado_2 = df_final.groupby(by=['ocor_MUNNOME', 'data_obito']).size().reset_index(name='obitos_totais')
```

```python
df_agrupado_2.head()
```

```python
# Vamos ordenar os dados por município e data de óbito
df_agrupado_2 = df_agrupado_2.sort_values(by=['ocor_MUNNOME', 'data_obito'])

# Em seguida iremos calcular os óbitos esperados
# não lembrava o valor a ser considerado para a média móvel então considerei em torno de um mês (30 dias)
df_agrupado_2['obitos_esperados'] = df_agrupado_2.groupby('ocor_MUNNOME')['obitos_totais'].transform(lambda x: x.rolling(30, min_periods=1).mean())

# Selecionar algumas cidades para análise
cidades_para_comparacao = ['Rio de Janeiro', 'São Paulo', 'Belo Horizonte']
df_comparacao = df_agrupado_2[df_agrupado_2['ocor_MUNNOME'].isin(cidades_para_comparacao)]

# Exibir os primeiros registros
print(df_comparacao.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo dos gráficos
sns.set_theme(style="darkgrid")

# Criar gráficos para cada cidade selecionada
cidades_para_comparacao = ['Rio de Janeiro', 'São Paulo', 'Belo Horizonte']

plt.figure(figsize=(12, 6))

for cidade in cidades_para_comparacao:
    df_cidade = df_comparacao[df_comparacao['ocor_MUNNOME'] == cidade]

    # Criar gráfico de linha
    plt.plot(df_cidade['data_obito'], df_cidade['obitos_totais'], label=f'{cidade} - Óbitos Totais', marker='o', linestyle='-')
    plt.plot(df_cidade['data_obito'], df_cidade['obitos_esperados'], label=f'{cidade} - Óbitos Esperados', marker='x', linestyle='--')

# Configurar rótulos e título
plt.xlabel('Data do Óbito')
plt.ylabel('Número de Óbitos')
plt.title('Óbitos Totais vs Esperados por Município')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Exibir gráfico
plt.show()
```
